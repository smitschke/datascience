```python
import numpy as np
import pandas as pd
import io
import requests
import time
from datetime import datetime
```

```python
url = "https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv"
request = requests.get(url).content

data = pd.read_csv(io.StringIO(request.decode('utf-8')))

data
```

```python
#Leere Datensätze entfernen
data.dropna(axis=0, how='all', inplace=True)
data
```

```python
#Doppelte Datensätze entfernen
data.drop_duplicates(subset=["full_name","first_name","last_name","email","gender","age"], inplace=True)
data
```

```python
#Leere Id's mit dem nächstgrößeren Wert füllen
#data.fillna(value={"id":data.id.max()+1}, inplace=True)

for x in range(data.id.isnull().sum()):
    data.fillna(value={"id":data.id.max()+1}, inplace=True, limit=1)
    #for-Schleife benötigt, damit aufsteigend gefüllt wird und nicht alle fehlenden Id's auf die gleiche, nächstgrößere Id (hier 22) gesetzt werden

#Float in Int umwandeln
data = data.astype({'id':'int32'})

#Index des Datenframes mit den Werten der Felder Id überschreiben
data.set_index("id", inplace=True)
#Falls die Id anschließend noch weiterverarbeitet werden soll ist zu beachten, dass diese nicht mehr als Column gelistet wird
#Zugriff über:
#for index, row in data.iterrows():
#    print(index, row["first_name"])
data
```

```python
#Fehlende Id's wurden bewusst im Dataframe belassen. Könnten aber mit folgender Zeile neu hochgezählt werden
data.index = pd.RangeIndex(start=1, stop=len(data.index)+1)
data
```

```python
#full_name nicht benötigt -> Normalform beachten
data = data.drop(['full_name'], axis=1)
data
```

```python
#Leere Email-Adressen herausfinden und überschreiben
data.fillna(value={"email":"undefined"}, inplace=True)

#Invalide Email-Adressen herausfinden und überschreiben
data.email.where(data["email"].str.match(r'^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$|^undefined$').astype(bool) == True, other="invalid", inplace=True)
data
```

```python
#Datentype der Einträge "age" von Object zu Numeric konvertieren. Errors="coerce" wandelt String in NaN um, die in der zweiten Zeile in 0 umgewandelt werden 
data["age"] = pd.to_numeric(data["age"], errors="coerce")
data.age.fillna(value=0, inplace=True)
#Nachdem innerhalb der Methode pd.to_numeric(downcast="integer") nicht funktioniert (warum eigentlich nicht?), hier nochmal eine separate Umwandlung in int und die Berechnung des Betrages, um negative Vorzeichen zu entfernen
data.age = abs(data.age.astype('int64'))

#Umrechnung Alter in Datum
for index, row in data.iterrows():
    if row["age"] != 0:
        yob = datetime.now().year-row["age"]
        #Umrechnung in Timestamp, falls Datum lieber in dem Format gespeicher werden soll
        #timestamp = pd.DatetimeIndex([datetime(yob, 1, 1)]).astype(np.int64)
        yob = pd.DatetimeIndex([datetime(yob, 1, 1)])
        data.at[index,"age"] = yob[0]

data
```

```python
#Setze weder männlich noch weibliche User auf "Diverse"
data.gender.fillna(value="Diverse", inplace=True)
data
```
